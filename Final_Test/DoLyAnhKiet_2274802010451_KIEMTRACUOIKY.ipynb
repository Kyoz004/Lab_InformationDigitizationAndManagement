### Câu 2. Điền giá trị thiếu trong Score bằng trung bình của Course tương ứng,

````python
def calculate_campus_course_score(df, campus, course):
    """Tính điểm trung bình cho từng campus và môn học"""
    scores = []
    for _, row in df.iterrows():
        if row['Campus'] == campus and row['Course'] == course and not pd.isnull(row['Score']):
            scores.append(row['Score'])
    return sum(scores) / len(scores) if scores else None

def fill_missing_scores(df):
    """Điền giá trị thiếu dựa trên đặc điểm của Campus"""
    df_filled = df.copy()
    campuses = df['Campus'].unique()
    courses = df['Course'].unique()
    
    # Tạo dictionary lưu trữ điểm trung bình
    campus_course_averages = {}
    for campus in campuses:
        for course in courses:
            avg = calculate_campus_course_score(df, campus, course)
            campus_course_averages[(campus, course)] = avg
    
    # Điền giá trị thiếu
    for idx, row in df_filled.iterrows():
        if pd.isnull(row['Score']):
            key = (row['Campus'], row['Course'])
            df_filled.at[idx, 'Score'] = campus_course_averages[key]
    
    return df_filled

# Áp dụng phương pháp điền giá trị
df_manual_filled = fill_missing_scores(df)

# Vẽ biểu đồ so sánh kết quả
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Dữ liệu gốc
ax1.hist(df['Score'].dropna(), bins=20, alpha=0.5, color='blue')
ax1.set_title('Phân phối điểm gốc')
ax1.set_xlabel('Điểm số')
ax1.set_ylabel('Tần suất')

# Dữ liệu sau khi điền
ax2.hist(df_manual_filled['Score'], bins=20, alpha=0.5, color='green')
ax2.set_title('Phân phối điểm sau khi điền')
ax2.set_xlabel('Điểm số')
ax2.set_ylabel('Tần suất')

plt.tight_layout()
plt.show()

# In thống kê so sánh
print("\nThống kê trước và sau khi điền giá trị:")
print("\nTrước:")
print(df['Score'].describe())
print("\nSau:")
print(df_manual_filled['Score'].describe())
````

````python
# Tiêu chí Xét duyệt
def evaluate_scholarship(student_data):
    """
    Đánh giá và phân loại học bổng
    """
    efficiency = student_data['Score'] / student_data['Study_Hours']
    if efficiency >= df['Efficiency'].quantile(0.95):
        return 'Platinum'
    elif efficiency >= df['Efficiency'].quantile(0.90):
        return 'Gold'
    elif efficiency >= df['Efficiency'].quantile(0.85):
        return 'Silver'
    return 'Not Eligible'
````

````python
# 1. Phương pháp Chuẩn hóa Z-score
def normalize_scores(scores):
    """
    Chuẩn hóa điểm số using Z-score
    """
    z_scores = (scores - scores.mean()) / scores.std()
    normalized = (z_scores * 10) + 75  # Scale to center around 75
    return np.clip(normalized, 0, 100)  # Clip to 0-100 range
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate high performers ratio by campus
def calculate_high_performers(df, threshold=80):
    campus_stats = df.groupby('Campus').agg({
        'Student_ID': 'count',  # total students
        'Score': lambda x: sum(x > threshold)  # students above threshold
    })
    
    campus_stats['high_performer_ratio'] = (
        campus_stats['Score'] / campus_stats['Student_ID'] * 100
    )
    return campus_stats

# Calculate and display results
campus_performance = calculate_high_performers(df)
print("\nTỷ lệ sinh viên điểm cao theo Campus:")
print(campus_performance['high_performer_ratio'])

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(campus_performance['high_performer_ratio'], 
        labels=campus_performance.index,
        autopct='%1.1f%%',
        colors=['lightblue', 'lightgreen', 'coral'])
plt.title('Tỷ lệ sinh viên điểm trên 80 theo Campus')
plt.show()
````

### 15. Tạo hàm Python xác định sinh viên có Score ngoài 2 độ lệch chuẩn

````python
import numpy as np
import pandas as pd

def identify_score_outliers(df, std_threshold=2):
    """
    Xác định sinh viên có điểm nằm ngoài khoảng ±2 độ lệch chuẩn
    
    Parameters:
        df: DataFrame chứa dữ liệu sinh viên
        std_threshold: Số độ lệch chuẩn (mặc định = 2)
    Returns:
        DataFrame chứa thông tin sinh viên outlier
    """
    # Tính thống kê cơ bản
    mean_score = df['Score'].mean()
    std_score = df['Score'].std()
    
    # Xác định ngưỡng
    lower_bound = mean_score - std_threshold * std_score
    upper_bound = mean_score + std_threshold * std_score
    
    # Lọc outliers
    outliers = df[
        (df['Score'] < lower_bound) | 
        (df['Score'] > upper_bound)
    ][['Student_ID', 'Course', 'Score', 'Attendance (%)', 'Study_Hours']]
    
    return outliers, lower_bound, upper_bound

# Áp dụng hàm
outliers, lower, upper = identify_score_outliers(df)
print(f"Ngưỡng điểm: {lower:.2f} - {upper:.2f}")
print("\nSinh viên có điểm ngoài ngưỡng:")
print(outliers)
````

````python
# 2. 3D Scatter Plot with Surface
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy.interpolate import griddata

# Create 3D surface plot for each course
def create_3d_surface(course_data):
    x = course_data['Attendance (%)']
    y = course_data['Study_Hours']
    z = course_data['Score']
    
    # Remove rows with NaN values
    valid_data = course_data.dropna(subset=['Attendance (%)', 'Study_Hours', 'Score'])
    x = valid_data['Attendance (%)']
    y = valid_data['Study_Hours']
    z = valid_data['Score']
    
    # Create grid for surface
    xi = np.linspace(x.min(), x.max(), 100)
    yi = np.linspace(y.min(), y.max(), 100)
    xi, yi = np.meshgrid(xi, yi)
    
    # Interpolate
    zi = griddata((x, y), z, (xi, yi), method='cubic')
    
    return go.Surface(x=xi, y=yi, z=zi, opacity=0.7)

# Create figure with subplots
fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=('Math', 'Physics', 'Chemistry'),
    specs=[[{'type': 'surface'}]*3]
)

# Add surfaces for each course
for i, course in enumerate(['Math', 'Physics', 'Chemistry'], 1):
    course_data = df[df['Course'] == course]
    fig.add_trace(
        create_3d_surface(course_data),
        row=1, col=i
    )

fig.update_layout(
    title='3D Score Surface by Course',
    height=800,
    scene=dict(
        xaxis_title='Attendance',
        yaxis_title='Study Hours',
        zaxis_title='Score'
    )
)

fig.show()
````

````python
# Đặc trưng Giáo dục Mới
def tao_dac_trung_giao_duc(df):
    """Tạo đặc trưng mới dựa trên ngữ cảnh giáo dục"""
    
    # Chỉ số tham gia học tập
    df['muc_do_tham_gia'] = (
        df['Attendance (%)'] / 100 * 
        df['Study_Hours'] / df['Study_Hours'].max()
    )
    
    # Điểm chuẩn hóa theo môn
    diem_trung_binh = df.groupby('Course')['Score'].transform('mean')
    df['hieu_suat_hoc_tap'] = df['Score'] / diem_trung_binh
    
    return df
````

````python
# Hàm Loss Theo Ngữ cảnh
class LossGiaoDuc(nn.Module):
    def __init__(self, alpha=0.3, beta=0.2):
        super(LossGiaoDuc, self).__init__()
        self.alpha = alpha  # Trọng số cho mức độ tham gia
        self.beta = beta   # Trọng số cho tính nhất quán
        
    def forward(self, du_doan, thuc_te, muc_do_tham_gia):
        # Loss cơ bản
        loss_co_ban = F.mse_loss(du_doan, thuc_te)
        
        # Loss có trọng số theo mức độ tham gia
        loss_trong_so = loss_co_ban * (1 + self.alpha * muc_do_tham_gia)
        
        # Phạt tính không nhất quán
        tinh_nhat_quan = torch.abs(du_doan[1:] - du_doan[:-1]).mean()
        
        return loss_trong_so.mean() + self.beta * tinh_nhat_quan
````

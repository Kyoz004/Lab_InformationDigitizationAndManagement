```python
import pandas as pd
import numpy as np

# D·ªØ li·ªáu ƒë·∫ßu v√†o
data = {
    "ID": [101, 102, 103, 104, 105],
    "NAME": ["Kiet", "Ben", None, "Kien", "Pino"],
    "AGE": [21, 22, 23, None, 25],
    "SALARY": [50000, 35000, 65000, 100000, None]
}

df = pd.DataFrame(data)

print("üîπ D·ªØ li·ªáu ban ƒë·∫ßu:")
print(df, "\n")

# 1Ô∏è‚É£ ƒêi·ªÅn gi√° tr·ªã thi·∫øu trong NAME b·∫±ng "Unknown" tr∆∞·ªõc khi x√≥a d·ªØ li·ªáu kh√°c
df["NAME"] = df["NAME"].fillna("Unknown")
print("üîπ Sau khi ƒëi·ªÅn gi√° tr·ªã thi·∫øu trong NAME b·∫±ng 'Unknown':")
print(df, "\n")

# 2Ô∏è‚É£ ƒêi·ªÅn gi√° tr·ªã thi·∫øu trong AGE b·∫±ng trung b√¨nh
df["AGE"] = df["AGE"].fillna(df["AGE"].mean())
print("üîπ Sau khi ƒëi·ªÅn gi√° tr·ªã thi·∫øu trong AGE b·∫±ng trung b√¨nh:")
print(df, "\n")

# 3Ô∏è‚É£ Ch·ªâ lo·∫°i b·ªè c√°c d√≤ng c√≥ d·ªØ li·ªáu thi·∫øu kh√¥ng th·ªÉ x·ª≠ l√Ω (SALARY = None)
df_cleaned = df.dropna(subset=["SALARY"]).copy()
print("üîπ Sau khi lo·∫°i b·ªè c√°c d√≤ng c√≥ SALARY b·ªã thi·∫øu:")
print(df_cleaned, "\n")

# 4Ô∏è‚É£ ƒêi·ªÅn gi√° tr·ªã thi·∫øu trong SALARY b·∫±ng ph∆∞∆°ng ph√°p n·ªôi suy (Interpolation)
df_cleaned["SALARY"] = df_cleaned["SALARY"].interpolate()
print("üîπ Sau khi n·ªôi suy gi√° tr·ªã thi·∫øu trong SALARY:")
print(df_cleaned, "\n")

# 5Ô∏è‚É£ In t·ª´ng gi√° tr·ªã sau khi x·ª≠ l√Ω
print("üîπ In t·ª´ng gi√° tr·ªã sau khi x·ª≠ l√Ω:")
for index, row in df_cleaned.iterrows():
    print(f"ID: {row['ID']}, NAME: {row['NAME']}, AGE: {row['AGE']}, SALARY: {row['SALARY']}")
```

```python
#cho data
import pandas as pd

df1 = pd.DataFrame({
    "Id": [1, 2, 3],
    "Score_A": [70, 90, 85]
})

df2 = pd.DataFrame({
    "Id": [3, 4, 5],
    "Score_A": [62, 91, 75]
})

# Th·ª±c hi·ªán c√°c ki·ªÉu merge tr√™n c·ªôt 'Id'
inter_join = pd.merge(df1, df2, on="Id", how="inner")  # Inner join
left_join = pd.merge(df1, df2, on="Id", how="left")    # Left join
outer_join = pd.merge(df1, df2, on="Id", how="outer")  # Outer join

# N·ªëi dataframe theo chi·ªÅu d·ªçc (vertical concatenation)
concat_df = pd.concat([df1, df2], ignore_index=True)

# G·ªôp df1 v√† df2 ƒë·ªÉ ƒëi·ªÅn gi√° tr·ªã thi·∫øu
merged_df = df1.set_index("Id").combine_first(df2.set_index("Id")).reset_index()

# In k·∫øt qu·∫£
print("Inter Join:\n", inter_join)
print("\nLeft Join:\n", left_join)
print("\nOuter Join:\n", outer_join)
print("\nConcatenated DataFrame:\n", concat_df)
print("\nMerged DataFrame (Filling Missing Values):\n", merged_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# D·ªØ li·ªáu
t = np.linspace(0, 10, 100)
y1 = np.sin(t)
y2 = np.cos(t)
y3 = np.sin(t) * np.cos(t)

# T·∫°o bi·ªÉu ƒë·ªì
plt.figure(figsize=(8, 5))  # K√≠ch th∆∞·ªõc bi·ªÉu ƒë·ªì

plt.plot(t, y1, label="sin(t)", color="blue", linestyle="-")    # ƒê∆∞·ªùng sin
plt.plot(t, y2, label="cos(t)", color="red", linestyle="--")    # ƒê∆∞·ªùng cos
plt.plot(t, y3, label="sin(t) * cos(t)", color="green", linestyle="-.")  # ƒê∆∞·ªùng sin * cos

# Th√™m ti√™u ƒë·ªÅ v√† nh√£n
plt.title("Bi·ªÉu ƒë·ªì nhi·ªÅu ƒë∆∞·ªùng theo th·ªùi gian")
plt.xlabel("Th·ªùi gian (t)")
plt.ylabel("Gi√° tr·ªã h√†m s·ªë")

# Hi·ªÉn th·ªã l∆∞·ªõi
plt.grid(True, linestyle=":")

# Hi·ªÉn th·ªã ch√∫ th√≠ch
plt.legend()

# Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# D·ªØ li·ªáu doanh thu c·ªßa A, B, C trong 4 qu√Ω
labels = ["Q1", "Q2", "Q3", "Q4"]
A = [500, 700, 800, 600]
B = [450, 350, 650, 750]
C = [500, 250, 850, 600]

# X√°c ƒë·ªãnh v·ªã tr√≠ c·ªßa c√°c nh√≥m c·ªôt
x = np.arange(len(labels))  # V·ªã tr√≠ c·ªßa nh√£n Q1, Q2, Q3, Q4
width = 0.25  # ƒê·ªô r·ªông c·ªßa m·ªói c·ªôt

# T·∫°o bi·ªÉu ƒë·ªì thanh nh√≥m
plt.figure(figsize=(8, 5))  # K√≠ch th∆∞·ªõc bi·ªÉu ƒë·ªì
plt.bar(x - width, A, width=width, label="A", color="blue")   # Doanh thu A
plt.bar(x, B, width=width, label="B", color="red")            # Doanh thu B
plt.bar(x + width, C, width=width, label="C", color="green")  # Doanh thu C

# Th√™m ti√™u ƒë·ªÅ v√† nh√£n
plt.title("Bi·ªÉu ƒë·ªì doanh thu c·ªßa A, B, C trong 4 qu√Ω")
plt.xlabel("Qu√Ω")
plt.ylabel("Doanh thu")
plt.xticks(x, labels)  # ƒê·∫∑t nh√£n tr·ª•c x l√† Q1, Q2, Q3, Q4

# Hi·ªÉn th·ªã gi√° tr·ªã tr√™n t·ª´ng c·ªôt
for i in range(len(labels)):
    plt.text(x[i] - width, A[i] + 10, str(A[i]), ha='center', fontsize=10)
    plt.text(x[i], B[i] + 10, str(B[i]), ha='center', fontsize=10)
    plt.text(x[i] + width, C[i] + 10, str(C[i]), ha='center', fontsize=10)

# Hi·ªÉn th·ªã ch√∫ th√≠ch
plt.legend()

# Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. Bi·ªÉu ƒë·ªì tr√≤n
Cty = ["A", "B", "C", "D"]
thiphan = [30, 25, 25, 20]
color = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]

plt.figure(figsize=(14, 4))

plt.subplot(1, 3, 1)
plt.pie(thiphan, labels=Cty, colors=color, autopct='%1.1f%%', startangle=140)
plt.title("Bi·ªÉu ƒë·ªì tr√≤n: T·ªâ l·ªá th·ªã ph·∫ßn c·ªßa 4 c√¥ng ty")

# 2. Bi·ªÉu ƒë·ªì ph√¢n t√°n
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100) * 300
colors = np.random.rand(100)

plt.subplot(1, 3, 2)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title("Bi·ªÉu ƒë·ªì ph√¢n t√°n c·ªßa hai bi·∫øn ng·∫´u nhi√™n")
plt.xlabel("Bi·∫øn X")
plt.ylabel("Bi·∫øn Y")

# 3. Bi·ªÉu ƒë·ªì nhi·ªát (heatmap)
data = np.random.rand(10, 10)

plt.subplot(1, 3, 3)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title("Heatmap: Ph√¢n b·ªë d·ªØ li·ªáu tr√™n ma tr·∫≠n")

plt.tight_layout()
plt.show()
```
